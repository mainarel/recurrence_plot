```python
import matplotlib.pyplot as plt
from pyts.image import RecurrencePlot
import numpy as np
```

# Тестовый пример

```python
times = np.arange(0,100,1)
datas = np.random.rand(100)
```

```python
time_points = np.linspace(0, 4 * np.pi, 1000)
x = np.sin(time_points)
#x = np.random.rand(500)
X = np.array([x])

# Recurrence plot transformation
rp = RecurrencePlot(threshold = 0.1)
X_rp = rp.fit_transform(X)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(X_rp[0], cmap='rainbow', origin='lower')
plt.title('Recurrence Plot')
plt.colorbar()
plt.show()
```

# Задача 1.

```python
num = np.loadtxt('/content/Haptics_TEST.txt', dtype=np.float)
```

```python
from scipy.io import arff
import pandas as pd

data = arff.loadarff('/content/Haptics_TEST.arff')
df = pd.DataFrame(data[0])
df.head(100)
```

```python
samples = np.array([(num[2])[1:], (num[11])[1:], (num[0])[1:], (num[1])[1:], (num[99])[1:]])
```

```python
plt.figure(figsize=(10, 5))
plt.grid()
plt.title("original dataset")
plt.ylabel("x-shift")
plt.plot((num[2])[1:], label = '1')
plt.plot((num[11])[1:], label = '2')
plt.plot((num[0])[1:], label = '3')
plt.plot((num[1])[1:], label = '4')
plt.plot((num[99])[1:], label = '5')
plt.legend()
```

```python
X = samples
rp = RecurrencePlot()
X_rp = rp.fit_transform(X)

for i in range (5):
     dd = plt.figure(i)
     ax = dd.add_subplot()
     print('Recurrence Plot для ', i+1, '-го человека')
     plt.imshow(X_rp[i], cmap='magma', origin='lower')
     plt.colorbar()
     plt.show()
```
